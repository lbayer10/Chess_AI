```python
import timeit
import chess

# for every piece on the board, make a 8x8 array
board = chess.Board()
board.push_san("e4")
```

```python
def board_to_booleans(board: chess.Board):
    booleans = []
    for color in chess.COLORS:
        # for every player, create an array of boards
        booleans.append([])
        for piece_type in chess.PIECE_TYPES:
            indexes = list(board.pieces(piece_type, color))
            # create the array for the current piece type
            array = [[False for x in range(8)] for y in range(8)]
            # for every index, set the array to true
            for index in indexes:
                array[index % 8][int(index / 8)] = True
            booleans[-1].append(array)
    return booleans

timeit.timeit(lambda: board_to_booleans(board), number=10000)
```

```python
# get amount of pieces on board
def get_piece_amount(board: chess.Board):
	piece_amount = 0
	for square in chess.SQUARES:
		if board.piece_at(square) is not None:
			piece_amount += 1
	return piece_amount
```

```python
fen = "r1bq1rk1/1pp2p2/p1np1n1p/2b1p1p1/B3P3/2PP1NB1/PP3PPP/RN1Q1RK1 b - - 0 10"
def create_board(board_str: str):
	board = chess.Board(board_str)
	return board


print(timeit.timeit(lambda: create_board(fen), number=10000))

print(timeit.timeit(lambda: create_board(chess.STARTING_FEN), number=10000))

print("Complex boards can take a lot longer to create ^")
```

```python
import numpy as np

game = np.load("memory/game-8406a1db.npy", allow_pickle=True)

print(f"Game length: {len(game)}")
print(f"Game result: {game[0][2]}")

game[0][0]

move_counter = len(game) - 1

print(f"Current game state: {game[move_counter][0]}")
moves = game[move_counter][1]
# get move with highest value from dictionary
best_move = max(moves, key=moves.get)
print(f"Best move: {best_move}")
print(len(moves), "moves:")
# sort moves
sorted_moves = sorted(moves.items(), key=lambda kv: kv[1], reverse=True)
for move in sorted_moves:
	print(f"{move[0]} : {move[1]}")
```
